# Scalling-Normalization

```python
import pandas as pd
```

```python
df=pd.read_csv('pima-indians-diabetes.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import normalize,scale
```

```python
X=df.drop('Outcome',axis=1)
```

```python
X_scale=scale(X)
```

```python
X_scale
```

```python
X_scale=pd.DataFrame(X_scale,columns=[X.columns])
```

```python
X_scale
```

```python
X_nrmlz=normalize(X)
```

```python
X_nrmlz
```

```python
X_nrmlz=pd.DataFrame(X_nrmlz,columns=[X.columns])
```

```python
X_nrmlz
```

### Boyut Küçültme PCA-Principle Component Analyse

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(2) #boyut sayısı
```

```python
df2=pca.fit_transform(X)
```

```python
df2
```

```python
X_pca=pd.DataFrame(df2,columns=['PCA1', 'PCA2'])
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_) # %95 i açıklıyor
```
